# 1. Load the Data and Analysis of Variables

```python
#Importing Important Libraries
import pandas as pd
```

```python
#Reading CSV File
df = pd.read_csv("/content/archive (6).zip")
```

```python
df.head()
```

# 2.Rename unwanted columns to make their use easier


```python
#Removing unwanted Column's
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
#Defining Dependent and Independent Variable
X = df.drop(["Survived"],axis="columns")
y = df.Survived
```

```python
#Get Dummy Values for Sex column
dummies = pd.get_dummies(X.Sex)
dummies.head()
```

```python
#Adding Dummies Column to the Dataframe
X = pd.concat([X,dummies],axis="columns")
X
```

```python
#Dropping Sex Column after getting Dummies Value for Male and Female
X.drop(["Sex"],axis="columns",inplace=True)
X.head()
```

# 3.check null values

```python
#Looking for Null Values in DataFrame
X.isna().sum()
```

```python
#Replacing Null Values with Mean of Age
X.Age = X.Age.fillna(X.Age.mean())
X.head()
```

```python
#Verifying that all Null Values are Gone
X.isna().sum()
```

# 4.Split test and train data set

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

# 5. Create model

```python
#Creating Naive Bayes Model
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
#Fitting the Model with Data
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.naive_bayes import GaussianNB

# Create an Imputer transformer
imputer = SimpleImputer(strategy='mean')  # You can also try 'median' or 'most_frequent' depending on your data

# Create a pipeline with the Imputer and the GaussianNB classifier
pipeline = make_pipeline(imputer, GaussianNB())

# Fit the pipeline on the training data
pipeline.fit(X_train, y_train)
```

```python
#Score of the Model
accuracy = pipeline.score(X_test, y_test)

print("Accuracy:", accuracy)
```

```python
#Predicting the Data of First 10 Passengers
predictions = pipeline.predict(X[:10])

print(predictions)
```
